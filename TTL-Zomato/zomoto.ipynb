```python
Country_values = [30, 20, 15]
Country_names = ['India', 'USA', 'China']

explode = (0.1, 0, 0)

colors = ['gold', 'green', 'blue']

plt.pie(Country_values, labels=Country_names, explode=explode, colors=colors, autopct='%1.2f%%', shadow=True, startangle=140)
plt.title('Distribution of Top 3 Countries')
plt.axis('equal')
plt.show()
```

```python
rating=final_df.groupby(["Aggregate rating","Rating color","Rating text"]).size().reset_index().rename(columns={0:"Rating Count"})
plt.figure(figsize=(12,4))
sns.barplot(x="Aggregate rating",y="Rating Count",data=rating,palette="viridis")
plt.xticks(rotation=90)
plt.title("RATINGS")
plt.xlabel("RATING")
plt.ylabel("COUNT")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x="Aggregate rating",y="Rating Count",hue="Rating color",data=rating,palette=["blue","red","orange","yellow","green","green"])
plt.show()
```
